```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
player_count = purchase_data["SN"].nunique()
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item Name"].nunique()
average_price = purchase_data["Price"].mean()
total_purchases = purchase_data["Purchase ID"].nunique()
total_revenue = purchase_data["Price"].sum()

purchase_summary_df = pd.DataFrame({"Number of Unique Item": [unique_items], 
                                "Average Price": [average_price],
                                 "Total Number of Purchases": [total_purchases],
                                 "Total Revenue": [total_revenue]})
purchase_summary_df
```

```python
#clean summary data frame to have cleaner data
purchase_summary_df["Average Price"] = purchase_summary_df["Average Price"].map('${:.2f}'.format)
purchase_summary_df["Total Revenue"] = purchase_summary_df["Total Revenue"].map('${:,}'.format)
purchase_summary_df
```

```python
gender_group = purchase_data.groupby(["Gender"])
gender_total_players = gender_group["SN"].nunique()
gender_total_percent = (gender_total_players / player_count) * 100

gender_demographics = pd.DataFrame({"Total Counts": gender_total_players,
                                   "Total Percentages": gender_total_percent})

gender_demographics
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_group_data = purchase_data.copy()
age_group_data
```

```python
max_age = purchase_data["Age"].max()
min_age = purchase_data["Age"].min()
(min_age, max_age)
```

```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 50]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
age_group_data["Age Group"] = pd.cut(age_group_data["Age"], bins, labels = labels, include_lowest = True, right = False)
age_group_data
```

```python
total = age_group_data["Age Group"].count()
total
```

```python
age_group = age_group_data.groupby(["Age Group"]).count()
age_demographic = pd.DataFrame(age_group)
age_demographic
```

```python
age_demographic_summary_df = age_demographic.loc[:, ["Age"]]
age_demographic_summary_df
```

```python
age_demographic_summary_df = age_demographic_summary_df.rename(columns={"Age": "Total Count"})
age_demographic_summary_df
```

```python
age_demographic_summary_df["Percent of Players"] = (age_demographic_summary_df['Total Count'] / total) * 100
age_demographic_summary_df
```

```python
#clean up age demographic summary table
age_demographic_summary_df["Percent of Players"] = age_demographic_summary_df["Percent of Players"].map('{:.2f}%'.format)
age_demographic_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
user_group = purchase_data.groupby(["SN"])
purchase_count = user_group["SN"].count()
total_purchase = user_group["Price"].sum()
average_purchase = user_group["Price"].mean()


top_spenders = pd.DataFrame({"Purchase Count": purchase_count,
                            "Average Purchase Price": average_purchase,
                            "Total Purchase Price": total_purchase})

top_spenders = top_spenders.sort_values(by="Total Purchase Price", ascending = False)
top_spenders
```

```python
#clean up top spenders table formatting, and sort total purchase count in descending order
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map('${:.2f}'.format)
top_spenders["Total Purchase Price"] = top_spenders["Total Purchase Price"].map('${:.2f}'.format)

top_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data
```

```python
most_popular = purchase_data.loc[:, ["Item ID", "Item Name", "Price"]]
most_popular = most_popular.groupby(["Item ID", "Item Name"])
```

```python
popular_total_purchase = most_popular["Price"].sum()
popular_average_purchase = most_popular["Price"].mean()
popular_purchase_count = most_popular["Price"].count()

most_popular_summary = pd.DataFrame({"Purchase Count": popular_purchase_count,
                               "Item Price": popular_average_purchase,
                               "Total Purchase Price": popular_total_purchase})
most_popular_summary = most_popular_summary.sort_values(by="Purchase Count", ascending = False)
most_popular_summary.head()
```

```python
#clean up most popular items table
most_popular_summary["Item Price"] = most_popular_summary["Item Price"].map('${:.2f}'.format)
most_popular_summary["Total Purchase Price"] = most_popular_summary["Total Purchase Price"].map('${:.2f}'.format)
most_popular_summary.head()
```
